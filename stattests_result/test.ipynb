```python
import pandas as pd
import os 
```

```python

def getListOfBestParams(medicine="SIB"):
    set_of_params = set()
    for folder in ["chisquare", "correlation", "ttest"]:
        for p in os.listdir(folder):
            if (p.find(medicine) != -1):
                f = pd.read_csv(os.path.join(folder, p), index_col=0)
                set_of_params.update(f.iloc[:, 0].values.tolist())
    with open(medicine+"_params.txt", "w", encoding="utf-8") as f:
        for p in list(set_of_params):
            f.write('"'+p+'"'+",\n")
getListOfBestParams("SIB")
getListOfBestParams("SAX")
getListOfBestParams("STD")
```

```python
m = {
    "SIB": "Сибутрамин (SIB)",
    "SAX": "Саксенда (SAX)",
    "STD": "Диета (STD)",
}
s = {
    "3percent": "потеря веса через 3 месяца",
    "6percent": "потеря веса через 6 месяца",
    "max3and6percent": "максимальная потеря веса",
}
abc = {
    "A": "потеря веса через 3 месяца",
    "B": "потеря веса через 6 месяца",
    "C": "максимальная потеря веса"
}
for i in ["3percent", "6percent", "max3and6percent"]:
    for j in ["SIB", "SAX", "STD"]:
        print(f"Корреляция: {m[j]} , {s[i]}")
        f = pd.read_csv(os.path.join("correlation", f"{j}_corr_{i}.csv"), index_col=0)
        display(f)
```

```python
for j in ["SIB", "SAX", "STD"]:
    print(f"Тесты Манни-Уитни и t-тест Стьюдента для {m[j]}. \n '+' и '-' mean и std - среднее и дисперсия параметра для пациентов с процентом потери веса выше и ниже порога \n count - количество значений")
    f = pd.read_csv(os.path.join("ttest", f"{j}_ttest.csv"), index_col=0)
    f = f[(f["Порог %"]==">5%") | (f["Порог %"]==">10%")]
    f = f[(f["MW p-value"] < 0.05) | (f["t-test p_value"] < 0.05)]
    f = f[f["Условие"].isin(["A", "B", "C"])]
    f["Условие"] = f["Условие"].map(abc)
    f.to_excel(f"{j}.xlsx")
    display(f)
```
